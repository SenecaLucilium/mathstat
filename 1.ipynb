```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
n = 100
tetta = 10

xi = tetta * np.random.random (n);
X = []
for i in range (1, n + 1):
    Xi = []
    for j in range (0, i):
        Xi.append (xi[j])
    X.append (Xi)
```

```python
def tetta1 (Xi):
    S = 0
    for i in Xi:
        S += i
    S = (2 / len(Xi)) * S
    return S

def tetta2 (Xi):
    return min(Xi)

def tetta2R (Xi):
    return min(Xi) * (len(Xi) + 1)

def tetta3 (Xi):
    return max(Xi)

def tetta3R (Xi):
    return max(Xi) * ((len(Xi) + 1) / len(Xi))

def tetta4 (Xi):
    return min(Xi) + max(Xi)

def tetta5 (Xi):
    S = Xi[0]
    for i in Xi[1:]:
        S += i / len(Xi)
    return S
```

```python
print (tetta1 (X[n - 1]))
print (tetta2 (X[n - 1]))
print (tetta3 (X[n - 1]))
print (tetta2R (X[n - 1]))
print (tetta3R (X[n - 1]))
print (tetta4 (X[n - 1]))
print (tetta5 (X[n - 1]))
```

```python
N = np.arange (1, n + 1)

T1 = []
for Xi in X:
    T1.append (np.abs (tetta1(Xi) - tetta))
T1 = np.array(T1)

T2 = []
for Xi in X:
    T2.append (np.abs (tetta2(Xi) - tetta))
T2 = np.array(T2)

T2R = []
for Xi in X:
    T2R.append (np.abs (tetta2R(Xi) - tetta))
T2R = np.array(T2R)

T3 = []
for Xi in X:
    T3.append (np.abs (tetta3(Xi) - tetta))
T3 = np.array(T3)

T3R = []
for Xi in X:
    T3R.append (np.abs (tetta3R(Xi) - tetta))
T3R = np.array(T3R)

T4 = []
for Xi in X:
    T4.append (np.abs (tetta4(Xi) - tetta))
T4 = np.array(T4)

T5 = []
for Xi in X:
    T5.append (np.abs (tetta5(Xi) - tetta))
T5 = np.array(T5)
```

```python
plt.figure (dpi=100)
plt.suptitle("Тетта 1")

plt.minorticks_on()
plt.grid(which='major', color = 'black', linewidth=0.8)
plt.grid(which='minor', color = 'lightgrey', linewidth=0.5)

plt.plot (N, T1, color='b')
```
